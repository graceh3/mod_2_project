## Adding dummy variables for county in df_ex-SI.csv dataframe

```python
import pandas as pd
import csv
```

```python
df = pd.read_csv('df_ex-SI.csv')
df.head()
```

```python
df = df.iloc[:, 2:]
df.head()
```

```python
df.dtypes
```

```python
cat_county = df.county.astype('category')
```

```python
cat_county.dtypes
```

```python
cat_county.cat.codes
```

```python
cat_dummies = pd.get_dummies(cat_county)
```

```python
df = pd.concat([df, cat_dummies], axis=1)
df.head()
```

```python
df.to_csv('df_ex-SI_dummies.csv')
```
