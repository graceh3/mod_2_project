```python
def make_neighborhood_lineplot(borough, neighborhood, df, x_label=None, y_label=None, title1=None):
    """generates a graph of median asking rent vs month from January 2010 to October 2018
    """
    borough = str(borough).title()
    neighborhood = str(neighborhood).title()
    plt.plot(np.linspace(1,105,num=105), df[df.Borough == borough][df.areaName == neighborhood][df.columns[4:]].T)
    if x_label != None:
        plt.xlabel(x_label)
    if y_label != None:
        plt.ylabel(y_label)
    if title1 != None:
        plt.title(title1)
```

```python
make_neighborhood_lineplot("Manhattan", "East Village", df_median_sales_price, x_label = 'months since Jan 2010', 
                           y_label = 'sales_price ($)', title1 = "East Village Sales Prices by Month")
```
