```python
zip_t2010 = nyc_geo[['borough','zipcode', 'tract2010']]
zip_t2010.head()
zip_t2010.tract2010 = zip_t2010.tract2010.map(lambda x: str(x))
zip_t2010['match_id'] = zip_t2010['borough'] +zip_t2010['tract2010']
zip_t2010.head()
```

```python
equivalency = pd.read_csv('2010_Census_Tract_to_Neighborhood_Tabulation_Area_Equivalency_table.csv')
equivalency.Borough.replace('Bronx', 'BX', inplace = True)
equivalency.Borough.replace('Queens', 'QN', inplace = True)
equivalency.Borough.replace('Brooklyn', 'BK', inplace = True)
equivalency.Borough.replace('Manhattan', 'MN', inplace = True)
equivalency.Borough.replace('Staten Island', 'SI', inplace = True)

equivalency.loc[(equivalency.Borough =='SI') & (equivalency['2010 Census Tract'] == 27702)]
```

```python
equivalency['2010 Census Tract'] = equivalency['2010 Census Tract'].map(lambda x: str(x)+'.0')
equivalency['match_id'] = equivalency['Borough'] + equivalency['2010 Census Tract']
equivalency.head()
```

```python
equivalency.loc[(equivalency.Borough =='SI') & (equivalency['2010 Census Tract'] == '27702.0')]
```

Merge zipcode & NTA codes

```python
zip_nta_ct2010 = zip_ct2010.merge(equivalency, how = 'inner', on ='match_id')
```

```python
zip_nta_ct2010.shape
```

```python
zip_nta_ct2010.rename(index=str, columns={"Neighborhood Tabulation Area (NTA)Code": "NTA_code", 
                                            "Neighborhood Tabulation Area (NTA) Name": "NTA_name"}, inplace = True)
zip_nta_ct2010.drop_duplicates(subset = 'NTA_code', inplace = True)
```
