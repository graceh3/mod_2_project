```python
import csv
```

```python
nyc_geo = pd.read_csv('pluto_18v2.csv')
```

```python
nyc_geo.ct2010.nunique()
```

```python
nyc_geo.zipcode.nunique()
```

```python
nyc_geo[['zipcode', 'ct2010']].head()
```

```python
zip_ct10 = nyc_geo[['borough','ct2010', 'zipcode']]
```

```python
zip_ct10.head()
```

```python
zip_ct10.ct2010 = zip_ct10.ct2010.map(lambda x: str(x))
zip_ct10['match_id'] = zip_ct10['borough'] + zip_ct10['ct2010']
zip_ct10.tail()
```

```python
zip_ct10.loc[(zip_ct10.borough =='BK') & (zip_ct10['ct2010'] == 1058.01)]
```

```python
equivalency = pd.read_csv('2010_Census_Tract_to_Neighborhood_Tabulation_Area_Equivalency_table.csv')
equivalency.Borough.replace('Bronx', 'BX', inplace = True)
equivalency.Borough.replace('Queens', 'QN', inplace = True)
equivalency.Borough.replace('Brooklyn', 'BK', inplace = True)
equivalency.Borough.replace('Manhattan', 'MN', inplace = True)
equivalency.Borough.replace('Staten Island', 'SI', inplace = True)
```

```python
equivalency['2010 Census Tract'] = equivalency['2010 Census Tract'].map(lambda x: str(x)[:-2]+'.'+str(x)[-1:])
equivalency['match_id'] = equivalency['Borough'] + equivalency['2010 Census Tract']
equivalency.head()
```

Merge zipcode & NTA codes

```python
df_key = zip_ct10.merge(equivalency, how = 'left', on ='match_id')
```

```python
df_key.shape
```

```python
df_key.rename(index=str, columns={"Neighborhood Tabulation Area (NTA)Code": "NTA_code", 
                                            "Neighborhood Tabulation Area (NTA) Name": "NTA_name"}, inplace = True)
```

```python
df_key.NTA_code.nunique()
```

```python
df_key.drop_duplicates('NTA_code', inplace=True)
```
